```python
# ğŸ“Š Dataset Analysis in Jupyter Notebook

# ØªØ­Ù…ÙŠÙ„ Ø§Ù„Ù…ÙƒØªØ¨Ø§Øª Ø§Ù„Ù„Ø§Ø²Ù…Ø©
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from fpdf import FPDF
from IPython.display import FileLink

# ğŸ“Œ 1ï¸âƒ£ ØªØ­Ù…ÙŠÙ„ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
df = pd.read_csv("data_science_extract.csv")  #  "your_dataset.csv" 
df.head()  # Ø¹Ø±Ø¶ Ø£ÙˆÙ„ 5 ØµÙÙˆÙ

# ğŸ“Œ 2ï¸âƒ£ ØªØ­Ù„ÙŠÙ„ Ø§Ù„Ù‚ÙŠÙ… Ø§Ù„Ù…ÙÙ‚ÙˆØ¯Ø©
missing_values = df.isnull().sum()
print("Missing Values per Column:")
print(missing_values[missing_values > 0])  # Ø¹Ø±Ø¶ Ø§Ù„Ø£Ø¹Ù…Ø¯Ø© Ø§Ù„ØªÙŠ ØªØ­ØªÙˆÙŠ Ø¹Ù„Ù‰ Ù‚ÙŠÙ… Ù…ÙÙ‚ÙˆØ¯Ø© ÙÙ‚Ø·

# ğŸ“Œ 3ï¸âƒ£ Ø¥Ø­ØµØ§Ø¦ÙŠØ§Øª Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„Ø¹Ø¯Ø¯ÙŠØ©
print("\nStatistical Summary:")
print(df.describe())  # Ø§Ø³ØªØ®Ø±Ø§Ø¬ Ø§Ù„Ø¥Ø­ØµØ§Ø¦ÙŠØ§Øª Ø§Ù„Ø£Ø³Ø§Ø³ÙŠØ© Ù…Ø«Ù„ Ø§Ù„Ù…ØªÙˆØ³Ø· ÙˆØ§Ù„ØªØ¨Ø§ÙŠÙ†

# ğŸ“Œ 4ï¸âƒ£ ØªÙˆØ²ÙŠØ¹ Ø¹Ø¯Ø¯ Ø§Ù„ÙƒÙ„Ù…Ø§Øª ÙÙŠ Ø§Ù„Ø£Ø¹Ù…Ø¯Ø© Ø§Ù„Ù†ØµÙŠØ©
column_name = "careerjunction_za_courses"  # Ø§Ø³ØªØ¨Ø¯Ù„ÙŠ Ø¨Ø§Ø³Ù… Ø§Ù„Ø¹Ù…ÙˆØ¯ Ø§Ù„Ù†ØµÙŠ Ø§Ù„Ø°ÙŠ ØªØ±ÙŠØ¯ÙŠÙ† ØªØ­Ù„ÙŠÙ„Ù‡
df[column_name] = df[column_name].astype(str)  # ØªØ­ÙˆÙŠÙ„ Ø§Ù„Ù‚ÙŠÙ… Ø¥Ù„Ù‰ Ù†Øµ Ù„ØªØ¬Ù†Ø¨ Ø§Ù„Ø£Ø®Ø·Ø§Ø¡
df["word_count"] = df[column_name].apply(lambda x: len(x.split()))

plt.figure(figsize=(8, 5))
sns.histplot(df["word_count"], bins=30, kde=True, color="blue")
plt.title(f"Word Count Distribution in {column_name}")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
plt.show()

# ğŸ“Œ 5ï¸âƒ£ Ø¥Ù†Ø´Ø§Ø¡ Ù…Ù„Ø®Øµ ÙˆØªØ­ÙˆÙŠÙ„Ù‡ Ø¥Ù„Ù‰ PDF
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)

summary_text = """
1ï¸âƒ£ Overview: Dataset contains job titles, skills, and courses.
2ï¸âƒ£ Distribution: Some columns have high unique values. No missing values found.
3ï¸âƒ£ Text Analysis: Word count varies significantly.
4ï¸âƒ£ Recommendations: Clean textual data and explore variable relationships.
"""

pdf.multi_cell(0, 10, summary_text)
pdf.output("Dataset_Summary.pdf")  # Ø­ÙØ¸ Ø§Ù„Ù…Ù„Ù

# ğŸ“Œ 6ï¸âƒ£ Ø¹Ø±Ø¶ Ø±Ø§Ø¨Ø· ØªØ­Ù…ÙŠÙ„ Ù…Ù„Ù Ø§Ù„Ù…Ù„Ø®Øµ
FileLink("Dataset_Summary.pdf")
```
